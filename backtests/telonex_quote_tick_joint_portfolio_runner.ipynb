# Telonex Quote-Tick Joint Portfolio Runner

Executes the fixed Telonex joint-portfolio runner and embeds the generated HTML summary report.

```python
from pathlib import Path
import sys

repo_root = Path(__file__).resolve().parents[1] if "__file__" in globals() else Path.cwd().resolve().parent
sys.path.insert(0, str(repo_root))

output_root = repo_root / "output"
output_root.mkdir(exist_ok=True)
```

```python
from backtests.polymarket_telonex_quote_tick_joint_portfolio_runner import EXPERIMENT, SUMMARY_REPORT_PATH
from prediction_market_extensions.backtesting._experiments import run_experiment_async
from prediction_market_extensions.backtesting._notebook_support import (
    display_html_artifacts,
    find_updated_html_artifacts,
    snapshot_html_artifacts,
)

html_snapshot = snapshot_html_artifacts(output_root)
results = await run_experiment_async(EXPERIMENT)
updated_html = find_updated_html_artifacts(output_root, html_snapshot)
summary_html = repo_root / SUMMARY_REPORT_PATH
html_artifacts = [*updated_html, *([] if summary_html in updated_html else [summary_html])]
display_html_artifacts([path for path in html_artifacts if path.exists()], repo_root=repo_root, iframe_height=900)
```
